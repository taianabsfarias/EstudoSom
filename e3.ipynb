```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
import librosa.display
```

```python
audio1, freq = librosa.load(r"C:\Users\taian\UFSC\LVA\teste2.waptt.wav", sr=None)
```

```python
Audio(data=audio1, rate=freq)
```

```python
plt.figure(figsize=(12,5))

audioEspec = librosa.amplitude_to_db(np.abs(librosa.stft(audio1)), ref=np.max)
# converte os sinais do dominio do tempo para o dominio da frequencia (amplitude_to_db) 
librosa.display.specshow(audioEspec, x_axis='time', y_axis='linear', sr=freq, cmap='jet')

plt.xlabel('Time [S]')
plt.ylabel('Frequency [Hz]')
plt.colorbar(format='%+2.0f dB')
plt.xticks(np.arange(0, len(audio1) / freq, step=1))
```

```python
fft = np.fft.fft(audio1)
# passa os sinais do dominio do tempo para o dominio da frequencia e cria um array com os valores
#  modulo dos valores  so as amplitudes de cada frequencia
freqfft = np.fft.fftfreq(len(audio1), d=1/freq)
# vetor com as frequencias de cada indice
n = len(audio1) // 2
# pega apenas as partes positivas
magnitude = np.abs(fft[:n])
fqfinal = freqfft[:n]

plt.plot(fqfinal, magnitude)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
