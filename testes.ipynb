```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
import librosa.display
from scipy.signal import butter, filtfilt
from scipy.fft import fft, fftfreq
```

```python
audioOriginal, taxaAmostragem = librosa.load(r"C:\Users\taian\UFSC\LVA\teste1.waptt.wav", sr=8000)
```

```python
Audio(data=audioOriginal, rate=taxaAmostragem)
```

```python
plt.figure(figsize=(12,5))

espectrodB = librosa.amplitude_to_db(np.abs(librosa.stft(audioOriginal)), ref=np.max)

librosa.display.specshow(espectrodB, x_axis='time', y_axis='linear', sr=taxaAmostragem, cmap='jet')

plt.xlabel('Time [S]')
plt.ylabel('Frequency [Hz]')
plt.colorbar(format='%+2.0f dB')
plt.xticks(np.arange(0, len(audioOriginal) / taxaAmostragem, step=1))
plt.show()
```
