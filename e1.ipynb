```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
import librosa.display
```

```python
audio1, freq = librosa.load(r"C:\Users\taian\UFSC\LVA\teste1.waptt.wav", sr=None)
```

```python
Audio(data=audio1, rate=freq)
```

```python
tempo = np.linspace(0, len(audio1) / freq, num=len(audio1))
#linspace cria um array a partir de (start, stop, num) (num  a quantidade de elementos dentro do array)
plt.figure(figsize=(12,5))
#plt.figure (plota uma figura) figsize(x,y) define o tamanho com as polegadas no eixo x e y
plt.plot(tempo, audio1)
# define os dados para o eixo x e o eixo y
plt.xlabel("Tempo [S]")
plt.ylabel("Amplitude")
#legendas dos eixos do grfico

plt.xticks(np.arange(0, len(audio1) / freq, step=0.5))
#plt.xticks controla a posio dos valores no eixo X do grafico
#np.arange(start, stop, step) cria um array com valores entre o start e o stop com um intervalo step entre eles

plt.grid()
plt.show()
```
