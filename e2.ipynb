```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
import librosa.display
```

```python
audio1, freq = librosa.load(r"C:\Users\taian\UFSC\LVA\teste2.waptt.wav", sr=None)
```

```python
Audio(data=audio1, rate=freq)
```

```python
audioNorm = audio1 / np.max(np.abs(audio1))
# todos os elementos do array ficam limitados entre -1 e 1 (normalização do sinal)
tempo = np.linspace(0, len(audio1) / freq, num=len(audio1))
# np.linspace é um array de valores espaçados (start, stop, step), nesse caso: (0, duração do audio, quantidade de amostras do audio)
# divide a quantidade de amostras pela taxa de amostragem (isso calcula a duração do audio em segundos)
plt.figure(figsize=(12,6))

plt.plot(tempo, audio1, color='blue', alpha=0.5)
plt.plot(tempo, audioNorm, color='red', alpha=0.5)

plt.xlabel("Tempo [S]")
plt.ylabel("Sinal normalizado")
plt.xticks(np.arange(0, len(audio1) / freq, step=0.5))
plt.grid()
plt.show()
```
